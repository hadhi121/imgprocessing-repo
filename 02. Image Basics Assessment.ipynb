### Import necessary libraries

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
```

### Open 'image2.jpg' from the data folder using open cv and display it in the notebook. Make sure to correct for the RGB order

```python
img = cv2.imread('image2.jpg')
```

##### Before RGB correction

```python
plt.imshow(img)
```

### After correction

```python
img_rgb = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Draw an empty RED rectangle around toms face and display the image in the notebook

```python
cv2.rectangle(img_rgb , pt1 = (300 , 400) , pt2 = (600 , 700) , color = (255 , 0 , 0) , thickness = 5)
plt.imshow(img_rgb)
```

### Draw a text named 'TOM' above toms face

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_rgb , text = 'TOM' , org = (500 , 200) ,fontFace = font , fontScale = 4 , color = (255 , 255 ,255) , thickness = 2 ,
           lineType = cv2.LINE_AA)
plt.imshow(img_rgb)
```

### Do the same steps for 'Jerry'

```python
font  = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_rgb , text = 'Jerry' ,  org = (130,350) , fontFace = font , fontScale = 4 , color = (255 , 255 , 255)  , thickness = 2 , 
            lineType = cv2.LINE_AA)
plt.imshow(img_rgb)
```

### Convert the image to grayscale

```python
img_gray = cv2.imread('image2.jpg' , cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray)
```

### Display the image in a larger size

```python
plt.figure(figsize = (8,8))
plt.imshow(img_rgb)
```
