# <span style = "color:#505AA8"> Blending and Pasting Images </span>

***

* Often we will be working with multiple images.
* OpenCV has many programming methods of blending images together and pasting images on top of each other.
* Blending images is done through the <b>addWeighted</b> function that uses both images and combines them.
* To blend images we use a simple formule:
<center><b> new_pixel = α x pixel_1 + β x pixel_2 + γ </b></center>

All your doing is a weighted sum.
If you want the image from pixel_1 to be more represented, Just make a higher 'α' value.

***

## Blending Images

Blending Images is actually quite simple, let's look at a simple example.

### Import Necessary Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read 'image3.jpg' and 'donotcopy.jpg'

Also make sure to convert these into into RGB format

```python
img1 = cv2.imread('image3.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('donotcopy.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

### Display both the images

```python
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()
```

#### Check shape of both images

```python
print(f"Shape of image 1 is {img1.shape}")
print(f"Shape of image 2 is {img2.shape}")
```

# Blending images of same size

### Resize both images to same shape

```python
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
```

### View images after the resize

```python
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()
```

### Blend the two images using addWeighted function in opencv

We will blend the values together with the formula:

$$  img1 * \alpha  + img2 * \beta  + \gamma $$

```python
blended = cv2.addWeighted(src1 = img1, alpha = 0.5, src2 = img2, beta = 0.5, gamma = 0)
plt.imshow(blended)
plt.show()
```

# Overlaying a small image on top of a larger image

We can use this trick to quickly overlap different sized images, by simply reassigning the larger image's values to match the smaller image.

### Read both the images once again

```python
img1 = cv2.imread('image3.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('donotcopy.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()
```

### Rename img1 to large_image and img2 to small_image for convienence

```python
large_image = img1
small_image = img2
```

### Set offset values

```python
x_offset = 0
y_offset = 0
```

### Set end values

```python
x_end = x_offset + small_image.shape[1]
y_end = y_offset + small_image.shape[0]
```

### Paste the smaller image on top of the larger image

```python
large_image[y_offset:y_end,x_offset:x_end] = small_image
```

### View the large image

```python
plt.imshow(large_image)
plt.show()
```
